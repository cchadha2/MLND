### Implementation: Define a Performance Metric
It is difficult to measure the quality of a given model without quantifying its performance over training and testing. This is typically done using some type of performance metric, whether it is through calculating some type of error, the goodness of fit, or some other useful measurement. For this project, you will be calculating the [*coefficient of determination*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination), R<sup>2</sup>, to quantify your model's performance. The coefficient of determination for a model is a useful statistic in regression analysis, as it often describes how "good" that model is at making predictions. 

The values for R<sup>2</sup> range from 0 to 1, which captures the percentage of squared correlation between the predicted and actual values of the **target variable**. A model with an R<sup>2</sup> of 0 is no better than a model that always predicts the *mean* of the target variable, whereas a model with an R<sup>2</sup> of 1 perfectly predicts the target variable. Any value between 0 and 1 indicates what percentage of the target variable, using this model, can be explained by the **features**. _A model can be given a negative R<sup>2</sup> as well, which indicates that the model is **arbitrarily worse** than one that always predicts the mean of the target variable._

For the `performance_metric` function in the code cell below, you will need to implement the following:
- Use `r2_score` from `sklearn.metrics` to perform a performance calculation between `y_true` and `y_predict`.
- Assign the performance score to the `score` variable.

```python
# TODO: Import 'r2_score'
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

**Hint:** Think about how overfitting or underfitting is contingent upon how splits on data is done.

**Answer: ** The benefit to splitting a dataset between training and testing subsets is that it provides the learning algorithm most of the dataset to train itself on (training set) and some of the dataset to validate its predictions upon (testing set). It is also advised to split the data into three subsets: training, cross-validation, and test. As before, the training set is used to train the model. In terms of the bias-variance tradeoff, splitting the dataset into training and testing sets is highly valuable. If we look at the model complexity graphs or learning curves of a range of models, splitting the dataset can easily show us which models are overfitting and which are underfitting. For example, as we increase the number of training points provided to the models, those that begin with low training error and high cross-validation error will see a gradual convergence into a similar training and CV error (relatively high). Similarly, an overfitting (high variance) model will begin with very low training error and very high CV error. Training error will remain quite low and CV error will reduce a small amount but remain high as the number of training points increase. This means that the model has basically memorised the data in the training set and does not generalise well (predictions will be inaccurate on data that is not the training set). However, the optimal model will have low training error and high CV error to begin but CV error will rapidly decrease as the number of training points increase and the two errors converge to relatively low amounts. This model will accurately reflect the complexity of the underlying data and should provide much better predictions than the other models. Without splitting the data, we would not be able to validate the model we create, effectively leaving us blind to whether it is underfitting or overfitting.

**Answer: 
** The grid search technique is a way of systematically working through multiple combinations of parameter tunes, while cross-validating each parameter combination, to determine which parameters provide the best performance based on a predetermined metric. GridSearchCV is a hyper-parameter optimiser in the sklearn.model_selection module. The user provides GridSearchCV a dictionary of parameters that they are looking to tune (each passed as an array of values including strings) and a metric to score each model using make_scorer from sklearn.metrics and a suitable metric (e.g. f1_score from sklearn.metrics). An object of GridSearchCV is then created using as parameters: the model we are looking to tune (with default parameters set) and the parameters we are looking to tune as well as the scoring metric. This object creates a grid of models, with each possible combination of parameters, trained on the training set. Each model is given a score (based on the metric previously chosen) using the cross-validation set. Finally, the model with the best score is chosen by fitting the object to our training data and using the .best_estimator_ function. For example, if we were looking to tune max_depth and min_samples_leaf of a decision tree classifier, we would create a dictionary such as: parameters = {'max_depth': [2, 3, 4, 5, 7, 9], 'min_samples_leaf': [1, 2, 4, 6, 8, 10]} and a scorer: scorer = make_scorer(f1_score). An object of the GridSearchCV class would then take the following shape: grid_obj = GridSearchCV(clf, parameters, scoring = scorer) where clf = sklearn.tree.DecisionTreeClassifier(). Note the empty parameters in the classifier. The GridSearchCV object is then fit to the training data: grid_fit = grid_obj.fit(X_train, y_train). This is a grid of decision trees made up of each possible combination of max_depth and min_samples_leaf as we presented in the dictionary, with all other parameters set to default (hence the empty parameters in classifier). From this, we can find our best estimator: best_clf = grid_fit.best_estimator_ as previously mentioned. Fitting best_clf to our data will give us the best predictions based on our chosen metric (F1 score).

### Question 8 - Cross-Validation

**Hint:** When explaining the k-fold cross validation technique, be sure to touch upon what 'k' is, how the dataset is split into different parts for training and testing and the number of times it is run based on the 'k' value.

**Answer: ** k-fold cross-validation is a method of cross-validating a model 'k' times and then taking the average of the cross-validation results as our model. Instead of simply splitting our dataset into some ratio of training and testing sets, it is split into 'k' sets or folds. The model is trained on the data in k-1 folds and tested on one fold. This is repeated k times with the fold used for testing changing each time. After k repititions, the model will have been tested on each fold once. An average of the results is then taken to build our model. The folds can also be stratified to include approximately the same percentage of samples of each target class as the complete set and can also be shuffled before divided into folds to decrease bias in the data. This technique provides a benefit for grid search when opimising a model as grid search previously hinged on simply using a binary split in our dataset to cross-validate and find optimal hyper-parameters. Essentially, training our model on a portion of data and testing on a smaller portion. This method may not take into account variations in the training and testing sets. However, with k-fold cross-validation, grid search is able to use all of the data to both train and test on which subsequently results in finer tuning of our chosen parameters.

### Implementation: Fitting a Model
Your final implementation requires that you bring everything together and train a model using the **decision tree algorithm**. To ensure that you are producing an optimized model, you will train the model using the grid search technique to optimize the `'max_depth'` parameter for the decision tree. The `'max_depth'` parameter can be thought of as how many questions the decision tree algorithm is allowed to ask about the data before making a prediction. Decision trees are part of a class of algorithms called *supervised learning algorithms*.

In addition, you will find your implementation is using `ShuffleSplit()` for an alternative form of cross-validation (see the `'cv_sets'` variable). While it is not the K-Fold cross-validation technique you describe in **Question 8**, this type of cross-validation technique is just as useful!. The `ShuffleSplit()` implementation below will create 10 (`'n_splits'`) shuffled sets, and for each shuffle, 20% (`'test_size'`) of the data will be used as the *validation set*. While you're working on your implementation, think about the contrasts and similarities it has to the K-fold cross-validation technique.

Please note that ShuffleSplit has different parameters in scikit-learn versions 0.17 and 0.18.
For the `fit_model` function in the code cell below, you will need to implement the following:
- Use [`DecisionTreeRegressor`](http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html) from `sklearn.tree` to create a decision tree regressor object.
  - Assign this object to the `'regressor'` variable.
- Create a dictionary for `'max_depth'` with the values from 1 to 10, and assign this to the `'params'` variable.
- Use [`make_scorer`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.make_scorer.html) from `sklearn.metrics` to create a scoring function object.
  - Pass the `performance_metric` function as a parameter to the object.
  - Assign this scoring function to the `'scoring_fnc'` variable.
- Use [`GridSearchCV`](http://scikit-learn.org/0.17/modules/generated/sklearn.grid_search.GridSearchCV.html) from `sklearn.grid_search` to create a grid search object.
  - Pass the variables `'regressor'`, `'params'`, `'scoring_fnc'`, and `'cv_sets'` as parameters to the object. 
  - Assign the `GridSearchCV` object to the `'grid'` variable.

```python
# TODO: Import 'make_scorer', 'DecisionTreeRegressor', and 'GridSearchCV'
from sklearn.metrics import make_scorer
from sklearn.tree import DecisionTreeRegressor
from sklearn.grid_search import GridSearchCV

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    
    # Create cross-validation sets from the training data
    # sklearn version 0.18: ShuffleSplit(n_splits=10, test_size=0.1, train_size=None, random_state=None)
    # sklearn versiin 0.17: ShuffleSplit(n, n_iter=10, test_size=0.1, train_size=None, random_state=None)
    #cv_sets = ShuffleSplit(X.shape[0], n_iter = 10, test_size = 0.20, random_state = 0)
    cv_sets = ShuffleSplit(n_splits=10, test_size=0.2, train_size=0.8, random_state=0)

    # TODO: Create a decision tree regressor object
    regressor = DecisionTreeRegressor()

    # TODO: Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    params = {'max_depth': np.array(range(1,11))}

    # TODO: Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc = make_scorer(performance_metric)

    # TODO: Create the grid search cv object --> GridSearchCV()
    # Make sure to include the right parameters in the object:
    # (estimator, param_grid, scoring, cv) which have values 'regressor', 'params', 'scoring_fnc', and 'cv_sets' respectively.
    grid = GridSearchCV(regressor, params, scoring_fnc)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)
    
    # Return the optimal model after fitting the data
    return grid.best_estimator_
```
